# スクレイピングの基本を確認
## ページのデータを全て取得

```python
# ●ヒント1
# スクレイピング可否は、ウェブサイトの利用規約を読んでから判断する
# robots.txtファイルを見ると、クローラーやスクレイピングを行う人に対しての禁止事項や間隔を何秒あければ良いか、指示が書かれている
# 詳細は別紙のPDFファイルを参照のこと
# JavaScriptで使われた動的に動くウェブサイトのスクレイピングは難易度が高く、BeautifulSoupではスクレイピング出来ない

# ●ヒント2
# BeautifulSoupはスクレイピングを行うモジュール
# Sleepは、for文を不用意に繰り返さないためのモジュールで、Sleepと言う文字通り、一定間隔を明けてfor文を実行することを指定する
```

```python
# ■ステップ2 ページ上の全ての要素を対象にスクレイピングを行う
# スクレイピング対象のURL
url = 'http://mensnote.boy.jp/document/home/index.html'   # 左記のURLを対象にスクレイピングを行う

# User-Agentを書く
headers = {'User-Agent' : 'Example Company, example@example.com'}   # スクレイピングを行うUser-Agentを知らせる

# 対象のコンテキストマネージャを実際にリクエストしている部分
re = urllib.request.Request(url, None, headers)
html = urllib.request.urlopen(re)

# 受け取ったサイトのコンテキストマネージャをhtmlにパース
soup = BeautifulSoup(html, 'html.parser')

# パース前
print(html)
print('\n=======ここから下がhtml=======') 

# パース後のhtmlの部分
print(soup)
```

```python
# ●ヒント
# 上記のソースが、スクレイピングを行う際の基本コードになる
# パースとは機械語を、ヒトが理解できる形式に変換することを指す

# ・class指定
# 任意の要素をスクレイピングする場合、class指定に注意する
# classはPythonでは予約語と呼ばれ、変数として使えないため、class_(アンダーバー)と変数名を変えてやる必要がある
# class指定しているのに、コードが実行されない場合、class_になっているかを確認する
```

## 目的の要素だけを取得

```python
# ■ステップ3 ページ上の目的の要素だけをスクレピングｓる
# findでh1を指定
print(soup.find('h1'))   # ページ上のh1タグの要素のみをスクレイピングする
```

```python
# ●ヒント
# ページに複数あるタグをスクレイピングする場合、soup.find(" ")関数で持ってこれるタグは、ページ上で出現する一番先頭のタグのみになる
# 例えば、ページ上のpタグが5つあった場合、(soup.find("p"))だと、ページ内に登場する一番先頭のpタグのみがスクレイピング対象になる
```

## 様々な要素の取得方法

```python
# ■ステップ4 ページ上の様々な要素を取得する
# ・ページタイトルを取得
print(soup.title, end = '\n\n')   # ページのタイトル要素をスクレイピングする

# ・pタグでp1クラスの要素を取得
print(soup.find('p', class_='p1'), end='\n\n')   # p1タグをスクレイピングする
# class_(アンダーバー)が付いている点が注意　classは既に別機能がアサインされている予約語に当たるため

# ・pタグでp1クラスの要素を全て取得
print(soup.find_all('p', class_='p1'), end='\n\n')   # ページ内のp1タグ全てをスクレイピングする

# ・pタグでp-id2のidの要素を取得
print(soup.find('p', id = 'p-id2'), end='\n\n')   # p-id2タグをスクレイピングする

# ・p1クラス直下のspan要素を取得
print(soup.select('.p1 > span'), end='\n\n')   # p1タグ以下のspanを選択しスクレイピングする

# ・contentクラス直下のulの下のliの１番目の要素を取得
print(soup.select('.content > ul > li')[0], end='\n\n')   # content以下のul、更にしたのli要素を選択し、スクレイピングする

# ・aタグの中のリンクを取得
print(soup.a.get('href'))   # aタグの中のリンクをスクレイピングする
```

```python
# ●ヒント
# ・soup.タグ名で、任意のタグを取得できる
# soup.titleの場合は、ページ上のtitle要素をスクレイピングすることが出来る
# '\n\n'は改行を表す

# ・soup.find_all関数
# soup.find_all関数を使うと、ページ上の任意の要素を全てスクレイピングすることが出来る
# タグがpで、classがp1の要素を取得できる

# ・soup.select関数
# 入れ子構造になっている要素を指定して、スクレイピングすることが出来る

# ・soup.a.get関数
# aタグのリンク内をスクレイピングすることが出来る
```

```python
# for文を使ったスクレイピング
a_list = soup.find_all('a')   # aタグを全て探し、a_listに代入する

for a in a_list:   # a_listに格納された要素に対して
    print(a.get('href'))   # aタグをスクレイピングする
```

```python
# ●ヒント
# for文を使うことで、aタグ全てをスクレイピングすることが出来る
```

```python
# wikipediaのコピーのページのリンクを取得
url = "http://mensnote.boy.jp/document/home/page1.html"   # 左記のURLをスクレイピング対象にし、urlに代入する
html = urllib.request.urlopen(url)
soup = BeautifulSoup(html, "html.parser")
link = soup.select("#header li li a")[:2]
page_list = ["http://mensnote.boy.jp/document/home/"+l.get("href") for l in link]
page_list   # page_listを出力する
```

```python
# pandas自動化バージョン
i = 0   # iを初期化する(0を代入する)
for p in page_list:   # page_listに含まれるpに対して

    dfs = pandas.read_html(p)   # pandasモジュールでpを読み込み、dfsに代入する
    s = 'pandas_sample' + str(i) + '.csv'   # pandas_sampleにi番目の文字列にcsvを追加し、sに代入する
    dfs[1].to_csv(s)   # dfsの1番目の要素をsとしてCSVファイルに書き出す
    i += 1   # iに1を追加する

    sleep(10)   # 10秒間、スリープする(for文を止める)
```

# jsonデータのスクレイピング

```python
# ■ステップ9 JSON形式のデータ取得
url5 = 'http://weather.livedoor.com/forecast/webservice/json/v1?city=130010'   # 左記のURLをスクレイピング対象とし、url5に代入する
response = urllib.request.urlopen(url5)
content = json.loads(response.read().decode('utf8'))   # jsonをロード、utf-8の文字コードでデコードし、contentに代入する
print(content)   # contentを出力する
```

```python
# ●ヒント
# ウェブ・サービスの中には、スクレイピングをしなくてもデータ提供するAPIという仕組みを持っている
# 多くのAPIで扱われる最も一般的なデータ形式がjsonである
# APIを使うメリットとして、データの加工が簡単であることが挙げられる
```
